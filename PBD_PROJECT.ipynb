   # KMeans Clustering on Wikipedia Articles


Step 1 :
        reading wikipedia articles from the folder

```python
import os
```

```python
Article=r"C:\Users\hp\OneDrive\Desktop\Article"

Articles_path=[os.path.join(Article, name) for name in os.listdir(Article)]

All_Articles=[]
```

```python
for x in Articles_path:
    with open(x, 'r', encoding="ISO-8859-1") as data:
        data_file=data.readlines()
        All_Articles.append(data_file)
```

```python
All_Articles
```

```python
Articles_in_String=' '.join(map(str,All_Articles))
```

```python
Articles_in_String
```

Step 2:
        Removing punctuation

```python
import re
After_Removing_Pun = re.sub(r'[^\w\s]', '', Articles_in_String)
After_Removing_Pun = re.sub('[^a-z A-Z]+', '', Articles_in_String)
After_Removing_Pun
```

Step 3: Removing Stopwords

```python
%pip install nltk
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
Data_in_text = word_tokenize (After_Removing_Pun)
Final_Articles = [word for word in Data_in_text if not word in stopwords.words()]
print(Final_Articles)
```

Step 4: Performing KMEANS CLUSTERING

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

```python
df = TfidfVectorizer(stop_words='english')
making_cluster = df.fit_transform(Final_Articles)
cluster=2
example = KMeans(n_clusters=cluster, init='k-means++', max_iter=100, n_init=1)
example.fit(making_cluster)
```

```python
print("Clusters-> ")
order_centroids = example.cluster_centers_.argsort()[:, ::-1]
xx = df.get_feature_names()
for i in range(cluster):
    print("Cluster %d:" % i),
    for j in order_centroids[i, :20]:
        print(' %s' % xx[j])
```
